```python
#Create column Num Children (number of children at home: small kids+teens)
customers_children = unique_df[['Kidhome','Teenhome']]
unique_df['Num Children'] = customers_children.sum(axis=1)
```

```python
#Create column Total Purchases containing the Total number of purchases made
customers_purchases = unique_df[['NumWebPurchases','NumCatalogPurchases','NumStorePurchases']]
unique_df['Total Purchases'] = customers_purchases.sum(axis=1)
```

```python
#Create column Total Spent containing the total amount spent on products
customers_amnt = unique_df[['MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts']]
unique_df['Total Spent'] = customers_amnt.sum(axis=1)
#Create a column containing the number of total offers accepted from Campaigns 1 to 5
customers_cmp = unique_df[['AcceptedCmp1','AcceptedCmp2','AcceptedCmp3','AcceptedCmp4','AcceptedCmp5']]
unique_df['Total Campaigns'] = customers_cmp.sum(axis=1)

unique_df.columns
```

```python
#Recency quantiles
unique_df['Recency'].quantile([0.25,0.5,0.75])
```

```python
#Categorize customers by recency
def cat_recency(n):
    if n == 0:
        return '5'
    elif n > 0 and n < 24:
        return '4'
    elif  n >= 24 and n < 49:
        return '3'
    elif n >= 49 and n < 74:
        return '2'
    elif n >= 74:
        return '1'
```

```python
#Total num of purchases made quantiles
unique_df['Total Purchases'].quantile([0.25,0.5,0.75])
```

```python
#Categorize customers by number of purchases

def cat_purchase(n):
    if n==0:
        return '1'
    elif n > 0 and n < 6:
        return '2'
    elif n >= 6 and n < 12:
        return '3'
    elif n >= 12 and n < 18:
        return '4'
    elif n >= 18:
        return '5'
```

```python
#Total Spent on products quantiles
unique_df['Total Spent'].quantile([0.25,0.5,0.75])
```

```python
#Categorize customers by total spend on products
def cat_spend(n):
    if n == 0:
        return '1'
    elif n > 0 and n <55:
        return '2'
    elif n >= 55 and n < 343:
        return '3'
    elif n >= 343 and n < 964:
        return '4'
    elif n >= 964:
        return '5'
```

```python
#Create column categorizing the following features

#Total Spent on products
unique_df['Spend Category'] = unique_df['Total Spent'].apply(cat_spend)

#Total number of purchases
unique_df['Purchases Category'] = unique_df['Total Purchases'].apply(cat_purchase)

#Number of days since the last purchase - Recency
unique_df['Recency Category'] = unique_df['Recency'].apply(cat_recency)
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Convert categorical columns to integers
unique_df['Recency Category'] = unique_df['Recency Category'].astype(int)
unique_df['Purchases Category'] = unique_df['Purchases Category'].astype(int)
unique_df['Spend Category'] = unique_df['Spend Category'].astype(int)

# Step 1: Extract the relevant columns
data = unique_df[['Recency Category', 'Purchases Category', 'Spend Category']]

# Step 2: Standardize the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Step 3: Determine the optimal number of clusters using the Elbow Method
inertia = []
k_values = range(1, 11)  # Test for 1 to 10 clusters

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_scaled)
    inertia.append(kmeans.inertia_)

# Plot the Elbow Curve
plt.figure(figsize=(8, 5))
plt.plot(k_values, inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.show()
```

```python
# Step 4: Perform K-Means Clustering with the chosen k (e.g., 3)
k = 3  # Choose the optimal k based on the elbow method
model = KMeans(n_clusters=k, random_state=42)
model.fit(data_scaled)
clusters = model.predict(data_scaled)

# Step 5: Add cluster labels to the DataFrame
unique_df['Clusters'] = clusters

# View the DataFrame with cluster assignments
print("Groups Assigned : \n")
unique_df[["Recency Category", "Purchases Category", "Spend Category", "Clusters"]]
```

```python
unique_df.to_csv('clustered_rfm_analysis.csv', index=False)
import matplotlib.pyplot as plt

# Create boxplots for the numerical columns grouped by 'Cluster'
unique_df.boxplot(
    column=['Recency Category', 'Purchases Category', 'Spend Category'],  # Columns for boxplot
    by='Clusters',  # Grouping by cluster
    layout=(2, 2),  # Layout of subplots
    figsize=(20, 15)  # Size of the figure
)

plt.suptitle('Boxplot of RFM Categories by Clusters')  # Title for the figure
plt.show()
```
